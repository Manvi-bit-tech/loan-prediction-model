```python
encoded_df = pd.get_dummies(df, prefix=None, prefix_sep='_', dummy_na=False, columns=[' education',' self_employed',],sparse=False,drop_first=False, dtype=None)
```

```python
#encoded_df = pd.get_dummies(df, prefix=None, prefix_sep='_', dummy_na=False, columns=[' education', ' self_employed',], sparse=False, drop_first=False, dtype=None) #convert object into int values
```

```python
encoded_df.head()
```

```python
encoded_df.drop(['loan_id'],axis=1,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_df[' loan_status'] = le.fit_transform(encoded_df[' loan_status'])
```

```python
x =encoded_df.drop(columns=[' loan_status']) #attributes
y=encoded_df[' loan_status'] #target
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
```
